```python
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# Beta conjugate prior

Let's imagine a coin. We have a weak belief that the coin is fair ($\theta=.5$). 

Let's model this prior as $Beta(2, 2)$:

```python
prior_success = prior_fail = 2

prior = stats.beta(prior_success, prior_fail)
```

Let's define our probability space and generate our PDF:

```python
# Define probability space
x = np.linspace(0, 1, 101)

# Get PDF
prior_pdf = prior.pdf(x)

# Plot PDF
plt.plot(x, prior_pdf, label='prior')
plt.xlabel('$\\theta$')
plt.ylabel('PDF')
plt.title('Distribution')
plt.legend()
plt.show()
```

Let's get new observations:

```python
new_obs = [1, 1, 1, 1, 1, 0, 0, 1, 1, 1, 1, 0, 0] 

new_success = np.sum(new_obs)
new_fail = len(new_obs) - new_success
```

...and model the likelihood and posterior:

```python
# Initialize distributions
likelihood = stats.beta(new_success, new_fail)
posterior = stats.beta(prior_success + new_success, prior_fail + new_fail)

# Compute PDFs
likelihood_pdf = likelihood.pdf(x)
posterior_pdf = posterior.pdf(x)
```

Let's plot all of them:

```python
# Plot PDF
plt.plot(x, prior_pdf, label='prior')
plt.plot(x, likelihood_pdf, label='likelihood')
plt.plot(x, posterior_pdf, label='posterior')
plt.xlabel('$\\theta$')
plt.ylabel('PDFs')
plt.title('Distribution')
plt.legend()
plt.show()
```
